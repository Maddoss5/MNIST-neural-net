```python
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split 
```

```python
from torch.autograd import Variable
```

```python
import torchvision.datasets as datasets
import torchvision.transforms as transforms
```

```python
train_data= datasets.MNIST(root='./data',train=True,transform=transforms.ToTensor())
```

```python
test_data= datasets.MNIST(root='./data',train=False,transform=transforms.ToTensor())
```

```python
train_data
```

```python
train_load=torch.utils.data.DataLoader(dataset=train_data,batch_size=1000,shuffle=True)
```

```python
test_load=torch.utils.data.DataLoader(dataset=test_data,batch_size=1000,shuffle=True)
```

```python
class neur(nn.Module):
    def __init__(self):
        super(neur,self).__init__()
        self.h1=nn.Linear(784,800)
        self.h2=nn.Linear(800,100)
        self.out=nn.Linear(100,10)
        self.sigm=nn.Sigmoid()
    def forward(self,x):
        x=self.sigm(self.h1(x))
        x=self.sigm(self.h2(x))
        x=self.sigm(self.out(x))
        
        return x
        
```

```python
net=neur()
```

```python
critereon=nn.CrossEntropyLoss()
```

```python
optimizer=torch.optim.Adam(net.parameters(),lr=0.011)
```

```python
for j in range(0,5):
    for i ,(img,label) in enumerate(train_load):
        images=Variable(img.view(-1,784))
        labels=Variable(label)
        optimizer.zero_grad()
        out=net.forward(images)
        loss=critereon(out,labels)
        loss.backward()
        optimizer.step()
        ha,predicted=torch.max(out.data,1)
        correct=((predicted==labels).sum())
        sze=labels.size(0)
        print(f"{j} number epoch , iteration number{i} loss is {loss} and accuracy is {100*correct/sze} %")
    
```

```python
testcorr=0
tot=0

for i,(img,label) in enumerate(test_load):
    image=Variable(img.view(-1,784))
    label=Variable(label)
    outs=net.forward(image)
    loss=critereon(outs,label)
    temp,pred=torch.max(outs.data,1)
    testcorr+=((label==pred).sum())
    tot+=labels.size(0)
```
